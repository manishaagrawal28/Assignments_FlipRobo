# sentiment analysis before & after corona virus outbreak

```python
#importing libraries
import nltk
import string

from nltk.tokenize import word_tokenize
from collections import Counter

import matplotlib.pyplot as plt
```

```python
#scrap the data from twitter before corona virus outbreak
def get_tweets():
    import GetOldTweets3 as got 
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch('Corona') \
        .setSince("2020-01-28") \
        .setUntil("2020-02-28") \
        .setMaxTweets(150)
    # Creation of list that contains all tweets
    tweets = got.manager.TweetManager.getTweets(tweetCriteria)
    # Creating list of chosen tweet data
    text_tweets_before = [[tweet.text] for tweet in tweets]
    return text_tweets_before
```

```python
# reading text file
text = ""
text_tweets_before = get_tweets()

length = len(text_tweets_before)

for i in range(0, length):
    text = text_tweets_before[i][0] + " " + text

# converting to lowercase
lower_case1 = text.lower()

# splitting text into words
#tokenized_words = cleaned_text.split()
```

```python
text_tweets_before
```

```python
from nltk.tokenize import regexp_tokenize

# Only numbers
regexp_tokenize(lower_case1,"[0-9]+")
```

```python
pos = nltk.pos_tag(nltk.word_tokenize(lower_case1))
pos
```

```python
# import stopwords
from nltk.corpus import stopwords

stop_words = set(nltk.corpus.stopwords.words('english'))
len(stop_words)
```

```python
string.punctuation

punct =string.punctuation
punct
```

```python
# Empty list to load clean data
cleaned_text1= []

#remove stopwords and punctuations
for word in nltk.word_tokenize(lower_case1):
    if word not in punct:
        if word not in stop_words:
            cleaned_text1.append(word)
    
print ('Original Length of text  == >', len(lower_case1))
print ('length of cleaned text ==>', len(cleaned_text1))
print ('\n',cleaned_text1 )
```

```python
from nltk.stem import WordNetLemmatizer

# Lemmatization

lemma_words1 = []

for word in cleaned_text1:
    word = WordNetLemmatizer().lemmatize(word)
    lemma_words1.append(word)
```

```python
# Get emotions text
emotion_list1 = []
with open('emotions.txt', 'r') as file:
    for line in file:
        clear_line = line.replace('\n', '').replace(',', '').replace("'", '').replace("/", '').replace("*", '').strip()
        word, emotion = clear_line.split(':')
        if word in lemma_words1:
            emotion_list1.append(emotion)

print("People emotions from the text \n", emotion_list1, '\n \n')


w1 = Counter(emotion_list1)
print("Count of each emotion \n", w1)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Test SentimentIntensityAnalyzer 
sia = SentimentIntensityAnalyzer()
```

```python
def sentiment_analyse(sentiment_text):
    
    score = SentimentIntensityAnalyzer().polarity_scores(sentiment_text)
    
    if score['neg'] > score['pos']:
        print("\n     ******Negative Sentiment*******")
        
    elif score['neg'] < score['pos']:
        print("\n     ******Positive Sentiment*******")
    else:
        print("Neutral Sentiment")
```

```python
#public sentiment before corona virus outbreak
sentiment_analyse(cleaned_text1)

fig, ax1 = plt.subplots()
ax1.bar(w1.keys(), w1.values())
fig.autofmt_xdate()
plt.savefig('before_outbreak.png')
plt.show()
```

```python
#public sentiment after corona virus outbreak
sentiment_analyse(cleaned_text2)

fig, ax1 = plt.subplots()
ax1.bar(w2.keys(), w2.values())
fig.autofmt_xdate()
plt.savefig('after_outbreak.png')
plt.show()
```

# Using Bag Of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
```

```python
#BagOfWords Approach
def text_matrix(A, countvect):
    terms_doc = countvect.fit_transform(A)
    return pd.DataFrame(terms_doc.toarray(),columns=countvect.get_feature_names())
```

```python
A = ['Coronaviruses are a group of related RNA viruses that cause diseases in mammals and birds',"In humans and birds, they cause respiratory tract infections that can range from mild to lethal"," Mild illnesses in humans include some cases of the common cold (which is also caused by other viruses, predominantly rhinoviruses), while more lethal varieties can cause SARS, MERS, and COVID-19"," In cows and pigs they cause diarrhea, while in mice they cause hepatitis and encephalomyelitis","There are as yet no vaccines or antiviral drugs to prevent or treat human coronavirus infections"]

c_vect_A= CountVectorizer()
print ("Below metrix is the Bag of Words approach")
text_matrix(A, c_vect_A)
```
